## 3. DATA PREPROCESSING
    . lower case
    . toknization
    . removing special charachters
    . removing stop words and punctuations
    . stemming

```python
import string
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def transofrm_text(text):
    # make the all text lower case
    text=text.lower() 

    # toknization
    text=nltk.word_tokenize(text)

    # remove special charachters only alpha & numerics remaining 
    modifiedText=[]
    
    for i in text:
        if (i.isalnum()):
            modifiedText.append(i)

    text=modifiedText[:]
    modifiedText.clear()

    # removing stop words and punctuations
    
    stop_words= stopwords.words('english')
    punctuations=string.punctuation
    
    for i in text:
        if (i not in stop_words and i not in punctuations):
            modifiedText.append(i)

    text=modifiedText[:]
    modifiedText.clear()

    # steming the words

    for i in text:
        modifiedText.append(ps.stem(i))
    
    return " ".join(modifiedText)
```

```python
df['transformed_text']=df['text'].apply(transofrm_text)
```

## 5. MODEL BUILIDNG 

```python
# convert text_transformed to vactroize text
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=CountVectorizer()
tfidf=TfidfVectorizer(max_features=2500)
```

```python
X=tfidf.fit_transform(df['transformed_text']).toarray()
Y=df['target'].values
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
# X=scaler.fit_transform(X)
```

```python
# split the data for trail and test
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.3,random_state=2)
```

```python
# now get the neive bayes model becauze our data is textual
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
y_predict_mnb=mnb.predict(x_test)
print(accuracy_score(y_test,y_predict_mnb))
print(precision_score(y_test,y_predict_mnb))
```
